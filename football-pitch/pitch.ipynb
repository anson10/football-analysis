```python
import matplotlib.pyplot as plt

from mplsoccer import Pitch, VerticalPitch
```

```python
# Basic Pitch
pitch = Pitch()
# specifying figure size (width, height)
fig, ax = pitch.draw(figsize=(8, 4))
```

```python
# Creating a football pitch
from mplsoccer.pitch import Pitch
pitch = Pitch(pitch_color='grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```

```python
# Different pitch setup
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(10, 8), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')

# y-axis-inverted: 0-80
# x-axis: 0 - 120
```

```python
# mplsoccer support 10 pitch types. It can be used by "pitch_type" arguement. 
pitch = Pitch(pitch_type='opta',  axis=True, label=True)
fig, ax = pitch.draw()
```

```python
# Pitch Orientations

#horizontal
pitch = Pitch(half=False)
fig, ax = pitch.draw()
```

```python
#vertical
pitch = VerticalPitch(half=False)
fig, ax = pitch.draw()
```

```python
# horizontal half
pitch = Pitch(half=True)
fig, ax = pitch.draw()
```

```python
# vertical half
pitch = VerticalPitch(half=True)
fig, ax = pitch.draw()
```

```python
# adding corner arcs
pitch = VerticalPitch(corner_arcs=True, half=True)
fig, ax = pitch.draw(figsize=(10, 7.727))
```

```python
# tactical lines
pitch = Pitch(positional=True, shade_middle=True, positional_color='#eadddd', shade_color='#f2f2f2')
fig, ax = pitch.draw()
```
